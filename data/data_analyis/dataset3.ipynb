```python
import json
import pandas as pd
import matplotlib.pyplot as plt

with open('../data_training/dataSet_3/requests') as f:
    data = f.readlines()

requests = []
for line in data:
    request = json.loads(line)
    requests.append(request)

df = pd.DataFrame(requests)

# Replace 'your_target_metaKey' with the specific metaKey you want to analyze
target_metaKey = "691e44923cf182f29b793be363ed27de32cdb274"

# Filter data for the specific metaKey using .loc method
target_data = df.loc[df["metaKey"] == target_metaKey]

# Extract the durationsInMs for the specific metaKey using the DataFrame column
durations_in_ms = target_data["durationsInMs"].tolist()

# Calculate statistics
mean_duration = target_data["durationsInMs"].mean()
median_duration = target_data["durationsInMs"].median()
min_duration = target_data["durationsInMs"].min()
max_duration = target_data["durationsInMs"].max()

# Create a histogram
plt.hist(durations_in_ms, bins=50, color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('Duration (ms)')
plt.ylabel('Frequency')
plt.title(f'Duration Distribution for metaKey: {target_metaKey}')

# Show the plot
plt.show()

# Print statistics
print(f"Mean Duration: {mean_duration} ms")
print(f"Median Duration: {median_duration} ms")
print(f"Minimum Duration: {min_duration} ms")
print(f"Maximum Duration: {max_duration} ms")

```

```python
import json
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

with open('../data_training/dataSet_3/requests') as f:
    data = f.readlines()

requests = []
for line in data:
    request = json.loads(line)
    requests.append(request)

df = pd.DataFrame(requests)

# Replace 'your_target_metaKey' with the specific metaKey you want to analyze
target_metaKey = "8fd13b83de53b911e072a5e8673a0a29da35c544"

# Filter data for the specific metaKey using .loc method
target_data = df.loc[df["metaKey"] == target_metaKey]

# Sort the data by startTime
sorted_data = target_data.sort_values(by="startTime")

# Convert Unix timestamp to datetime
sorted_data["datetime"] = pd.to_datetime(sorted_data["startTime"], unit='ms')

# Extract the startTime and durationsInMs for the specific metaKey
start_times = sorted_data["datetime"]
durations_in_ms = sorted_data["durationsInMs"]

# Plot durationsInMs against startTime
plt.plot(start_times, durations_in_ms, marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Date and Time')
plt.ylabel('Duration (ms)')
plt.title(f'Duration over Time for metaKey: {target_metaKey}')

# Format the x-axis labels with both date and time
date_format = mdates.DateFormatter('%Y-%m-%d %H:%M:%S')
plt.gca().xaxis.set_major_formatter(date_format)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
